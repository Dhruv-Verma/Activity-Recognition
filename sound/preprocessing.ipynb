```python
import numpy as np
import resampy
import os
import matplotlib.pyplot as plt
import pickle
import vggish_input
import pickle as pkl
```

```python
data_path = os.path.join(os.getcwd(), 'ubicoustics_data/test')

class_labels = os.listdir(data_path)
len(class_labels)
```

```python
scale = 1
sr = 16000//scale
X, y = [], []
sr
```

```python
class_labels
```

```python
for scale in [1,8,16,32,64,128]:
    sr = 16000//scale
    X, y = [], []
    for label in class_labels:
        print('\n------{}------'.format(label)) 
        class_path = os.path.join(data_path, label)
        all_files = os.listdir(class_path)
        all_files = [file for file in all_files if not file.startswith('.')]

        files = all_files

        d=[]
        a=[]
        for num, file in enumerate(files):
            try:          
                file_path = os.path.join(class_path, file)
                file_features = vggish_input.wavfile_to_examples(file_path, scale)
                d.append(file_features)
                file_labels = [label] * len(file_features)
                a.append(file_labels)
                print('Completed {}/{} files in {}-{}'.format(num+1, len(files), label, 'train'))
            except:
                print('Error in opening file {} '.format(file))

    #     d = np.vstack(d)
    #     a = [label] * len(d)
        X.append(d)
        y.append(a)

    pkl.dump(X, open('featurized_data/downsample/X_test_{}.pkl'.format(sr), 'wb'))
    pkl.dump(y, open('featurized_data/downsample/y_test_{}.pkl'.format(sr), 'wb'))
```
